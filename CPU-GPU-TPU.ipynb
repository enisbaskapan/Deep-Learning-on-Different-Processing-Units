## Time Increase Rate in TPU vs GPU

It seems that for batch size 32, both CPU and GPU training times increase at the same rate. <br>
**Example:** From Epoch 1 to Epoch 5 ---> TPU 4 times more, GPU about 4 times more

For higher batch sizes and longer epochs, TPU time increase is shorter. <br>
**Example:** From Epoch 5 to Epoch 10 for batch size 512 ---> TPU is 1.5 times more but GPU is 2 times more <br>
**Example:** From Epoch 5 to Epoch 10 for batch size 1536 --> TPU is 1.3 times more but GPU is about 2 times more

From this conclusion, we can say that as the batch size and epoch increase, using TPU might be more efficient and it might become close to or faster than GPU speed. <br>
Dataset size and model complexity also effects this.
